# Reviewing Duplicate Columns 

```python

# List of duplicate columns
duplicate_columns = {
    'country': ['country_x', 'country_y'],
    'population': ['population_x', 'population_y'],
    'land_area': ['land_area_km2', 'area_km2']
}

print("Duplicate Columns Identified:")
for key, cols in duplicate_columns.items():
    print(f"{key}: {cols}")
```

# Comparing Duplicate Columns

```python
# Compare Duplicate dataframe_columns countries

# Compare 'country_x' and 'country_y'
country_match = country_df['country_x'].equals(country_df['country_y'])
print(f"Are 'country_x' and 'country_y' identical? {country_match}")

# If not identical, display mismatches
if not country_match:
    mismatches = country_df[country_df['country_x'] != country_df['country_y']]
    print("Mismatched Countries:")
    print(mismatches[['country_x', 'country_y']].head())
```

```python
# Compare duplicates ddataframe_columns populations 

# Compare 'population_x' and 'population_y'
population_match = country_df['population_x'].equals(country_df['population_y'])
print(f"Are 'population_x' and 'population_y' identical? {population_match}")

# If not identical, display some differences
if not population_match:
    population_diff = country_df[country_df['population_x'] != country_df['population_y']]
    print("Differences in Population:")
    print(population_diff[['country_x', 'population_x', 'population_y']].head())
```

```python
# Compare duplicates land area

# Compare 'land_area_km2' and 'area_km2'
land_area_match = country_df['land_area_km2'].equals(country_df['area_km2'])
print(f"Are 'land_area_km2' and 'area_km2' identical? {land_area_match}")

# If not identical, display some differences
if not land_area_match:
    land_area_diff = country_df[country_df['land_area_km2'] != country_df['area_km2']]
    print("Differences in Land Area:")
    print(land_area_diff[['country_x', 'land_area_km2', 'area_km2']].head())
```

*after doing research **area_km2** is the total area of the country (including water bodies)*

# Deciding which Columns to Keep

```python
# Handling country differences 
# Drop 'country_y' if 'country_x' is preferred
country_df = country_df.drop(columns=['country_y'])

# Rename 'country_x' to 'country'
country_df = country_df.rename(columns={'country_x': 'country'})
```

```python
# Handling population differences 

# Create 'average_population' column
country_df['average_population'] = country_df[['population_x', 'population_y']].mean(axis=1)

# Optionally, round the population to integers
country_df['average_population'] = country_df['average_population'].round(0).astype(int)

# Drop the original population columns
country_df = country_df.drop(columns=['population_x', 'population_y'])
```

```python
# rename area to total area for clarification
country_df = country_df.rename(columns={'area_km2': 'total_area_km2'})
```

# Verifying Changes

```python
# Checking changes made
# Display updated column names
print("Updated Columns:")
print(country_df.columns.tolist())

# Display the first few rows to verify changes
country_df.head()
```

# FInal Cleanup and Standarization

```python
# Standardize column names
country_df.columns = country_df.columns.str.lower().str.replace(' ', '_').str.replace('-', '_')

print("Standardized Column Names:")
print(country_df.columns.tolist())
```
